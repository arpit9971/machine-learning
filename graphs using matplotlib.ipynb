```python
import matplotlib.pyplot as plt
activities=['eat','sleep','work','play','dance','singing']
slices=[2,5,3,7,9,4]
plt.pie(slices,labels=activities,startangle=90,shadow=True,explode=(0,0,0.4,0,0,0),autopct='%1.2f%%')
#shadow is used for 3-D effect.
#explode is used to show the particular section out of the chart.
#autopct is used to calculate the percentage of each section.
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0,2*np.pi,0.01)
y1=np.sin(x)
y2=np.cos(x)
plt.plot(x,y1,label='sin function')
plt.plot(x,y2,label='cos function')
plt.legend()
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("sin and cos function")
plt.show()
```
